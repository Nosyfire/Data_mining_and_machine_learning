```python
X_train_digits, X_test_digits, y_train_digits, y_test_digits = train_test_split(digits.data, digits.target, train_size=0.7)
```

```python
def sigmoid(X:np.array):
        return 1/(1+np.exp(-X))

def accuracy(y_pred,y_test):
    return np.sum(y_pred==y_test) / len(y_test)

class LogisticRegression():

    def __init__(self, lr: float = 0.001, n_iters: int = 1000):
        """
        :param lr: learning rate
        :param n_iters: number of iteration
        """
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X: np.array, y: np.array):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias
            predictions = sigmoid(linear_pred)

            dw = (1/n_samples) * np.dot(X.T,(predictions - y))
            db = (1/n_samples) * np.sum(predictions - y)

            self.weights = self.weights - self.lr * dw
            self.bias = self.bias - self.lr * db

    def predict_proba(self, X):
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        self.probability = y_pred
        return np.array(y_pred)

    def predict(self, X):
        class_pred = [0 if y <= 0.5 else 1 for y in self.predict_proba(X)]
        return np.array(class_pred)
```

```python
logisticr = LogisticRegression(lr=0.00001)
logisticr.fit(X_train_digits,y_train_digits)
y_pred_digits = logisticr.predict(X_test_digits)
y_pred_digits_proba = logisticr.probability
```

```python
print("Weight is : " ,logisticr.weights)
print("Accuracy is : " ,accuracy(y_pred_digits,y_test_digits))
print("Bias is : " ,logisticr.bias)
```

```python
fpr, tpr, _ = roc_curve(y_test_digits,  y_pred_digits_proba)
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
%%time
lrates = np.linspace( 1e-5, 1, 10 )
acc_lrates = np.empty( lrates.shape[0], dtype=float )
for i in range(lrates.shape[0]):
    logisticr = LogisticRegression(lr=lrates[i])
    logisticr.fit(X_train_digits,y_train_digits)
    y_pred_digits = logisticr.predict(X_test_digits)
    acc_lrates[i] = accuracy(y_pred_digits,y_test_digits)
lrates_best = lrates[ np.argmax( acc_lrates, axis=0 ) ]

fig = plt.figure( figsize=(6,4), dpi=100 )
plt.title( 'Tuning the hyperparameters of LogisticsRegression', fontsize=20)
plt.plot( lrates, acc_lrates, label='Accuracy' )
plt.ylabel( 'Accuracy', fontsize = 15)
plt.xlabel( 'Learning rates', fontsize = 15)
plt.xscale( 'log' )
plt.grid()
plt.legend()
```

```python
%%time
iterations = np.linspace( 10, 10000, 10 )
acc_iters = np.empty( iterations.shape[0], dtype=float )
for i in range(iterations.shape[0]):
    logisticr = LogisticRegression(n_iters=int(iterations[i]))
    logisticr.fit(X_train_digits,y_train_digits)
    y_pred_digits = logisticr.predict(X_test_digits)
    acc_iters[i] = accuracy(y_pred_digits,y_test_digits)
n_iterations_best = iterations[ np.argmax( acc_iters, axis=0 ) ]

fig = plt.figure( figsize=(6,4), dpi=100 )
plt.title( 'Tuning the hyperparameters of Logistic Regression ', fontsize=20)
plt.plot( iterations, acc_iters, label='Accuracy' )
plt.ylabel( 'accuracy value', fontsize = 15)
plt.xlabel( 'Number of iteration', fontsize = 15)
plt.xscale( 'log' )
plt.grid()
plt.legend()
```

```python
# load mnist dataset from keras library
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

print('x_train shape: ', x_train.shape)
print('y_train shape: ', y_train.shape)
print('x_test shape: ', x_test.shape)
print('y_test shape: ', y_test.shape)
```

```python
plt.imshow(x_train[0], cmap=plt.cm.binary)
```

```python
y_train[0]
```

```python
# Normalize and reshape in one step
x_train_norm = x_train.reshape(60000, 28*28)/255
x_test_norm = x_test.reshape(10000, 28*28)/255
```

```python
x_train_norm[:5]
```

```python
y_train[:5]
```

```python
# Convert the labels to one-hot encoded form
class_number = 10 # label range: 0-9
y_train_oh = keras.utils.to_categorical(y_train, num_classes=class_number)
y_test_oh = keras.utils.to_categorical(y_test, num_classes=class_number)
```

```python
y_train_oh[:5]
```

```python
# Build the model
seq_model = keras.models.Sequential()

seq_model.add(keras.layers.Dense(750, activation='relu', input_dim=784))
seq_model.add(keras.layers.Dense(500, activation='relu'))
seq_model.add(keras.layers.Dense(500, activation='relu'))
seq_model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
seq_model.compile(optimizer=keras.optimizers.SGD(lr=1e-2), loss='categorical_crossentropy', metrics=['accuracy', 'AUC'])
```

```python
seq_model_history = seq_model.fit(x=x_train_norm, y=y_train_oh, batch_size=64, epochs=10)
```
